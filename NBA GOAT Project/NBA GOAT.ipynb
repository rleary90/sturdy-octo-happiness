# Create For Loop for all the BAA years, then aba years, then all nba years

```python
yearsBAA = list(range(1947,1950))
yearsABA = list(range(1968,1977))
yearsNBA = list(range(1950,2020))
yearsBAA
```

```python
statsBAA = {}
for yearBAA in yearsBAA:
    # URL page we will scraping (see image above)
    url = "https://www.basketball-reference.com/leagues/BAA_{}_advanced.html".format(yearBAA)
    # this is the HTML from the given URL
    html = urlopen(url)
    soup = BeautifulSoup(html)
    # use findALL() to get the column headers
    soup.findAll('tr', limit=2)
    # use getText()to extract the text we need into a list
    headers = [th.getText() for th in soup.findAll('tr', limit=2)[0].findAll('th')]
    # exclude the first column as we will not need the ranking order from Basketball Reference for the analysis
    headers = headers[1:]
    headers
    # avoid the first header row
    rows = soup.findAll('tr')[1:]
    player_stats = [[td.getText() for td in rows[i].findAll('td')]
            for i in range(len(rows))]
    
    statsBAA[yearBAA] = pd.DataFrame(player_stats, columns = headers)
    statsBAA[yearBAA]['League'] = 'BAA'
```

```python
for k in statsBAA:
    statsBAA[k].insert(0, 'key', [k]*len(statsBAA[k]))

statsBAA_df = pd.concat(statsBAA.values())
display(statsBAA_df)
```

```python
statsABA = {}
for yearABA in yearsABA:    
    # URL page we will scraping (see image above)
    url = "https://www.basketball-reference.com/leagues/ABA_{}_advanced.html".format(yearABA)
    # this is the HTML from the given URL
    html = urlopen(url)
    soup = BeautifulSoup(html)
    # use findALL() to get the column headers
    soup.findAll('tr', limit=2)
    # use getText()to extract the text we need into a list
    headers = [th.getText() for th in soup.findAll('tr', limit=2)[0].findAll('th')]
    # exclude the first column as we will not need the ranking order from Basketball Reference for the analysis
    headers = headers[1:]
    headers
    # avoid the first header row
    rows = soup.findAll('tr')[1:]
    player_stats = [[td.getText() for td in rows[i].findAll('td')]
            for i in range(len(rows))]
    statsABA[yearABA] = pd.DataFrame(player_stats, columns = headers)
    statsABA[yearABA]['League'] = 'ABA'
```

```python
for k in statsABA:
    statsABA[k].insert(0, 'key', [k]*len(statsABA[k]))

statsABA_df = pd.concat(statsABA.values())
display(statsABA_df)
```

```python
statsNBA = {}
for yearNBA in yearsNBA:    
    # URL page we will scraping (see image above)
    url = "https://www.basketball-reference.com/leagues/NBA_{}_advanced.html".format(yearNBA)
    # this is the HTML from the given URL
    html = urlopen(url)
    soup = BeautifulSoup(html)
    # use findALL() to get the column headers
    soup.findAll('tr', limit=2)
    # use getText()to extract the text we need into a list
    headers = [th.getText() for th in soup.findAll('tr', limit=2)[0].findAll('th')]
    # exclude the first column as we will not need the ranking order from Basketball Reference for the analysis
    headers = headers[1:]
    headers
    # avoid the first header row
    rows = soup.findAll('tr')[1:]
    player_stats = [[td.getText() for td in rows[i].findAll('td')]
            for i in range(len(rows))]
    statsNBA[yearNBA] = pd.DataFrame(player_stats, columns = headers)
    statsNBA[yearNBA]['League'] = 'NBA'
```

```python
for k in statsNBA:
    statsNBA[k].insert(0, 'key', [k]*len(statsNBA[k]))

statsNBA_df = pd.concat(statsNBA.values())
display(statsNBA_df)
```

```python
frames = [statsBAA_df, statsABA_df, statsNBA_df]
stats_df = pd.concat(frames, ignore_index=True)
```

```python
stats_df = stats_df.drop_duplicates(['key','Player'],keep= 'first')
```

```python
stats_df = stats_df.reset_index(drop=True)
```

```python
stats_df
```

```python
stats_df.info()
```

```python
stats_df = stats_df.apply(pd.to_numeric, errors='ignore')
```

```python
stats_df.info()
```

```python
stats_df['Year of Birth'] = stats_df['key'] - stats_df['Age']
stats_df
```

```python
stats_df
```

```python
stats_df = stats_df.loc[:,~stats_df.columns.duplicated()]
```

```python
career_stats = stats_df.groupby(['Player', 'Year of Birth']).agg({'G':'sum','WS':'sum'}).reset_index()
```

```python
career_stats = career_stats.sort_values(by='WS', ascending=False)
career_stats
```

```python
career_stats['Seasons'] = career_stats['G']/82
career_stats
```

```python
career_stats['WS/Season'] = career_stats['WS']/career_stats['Seasons']
career_stats
```

```python
career_stats['Time line adjusted'] = (career_stats['Year of Birth'] - 1800)/10
```

```python
career_stats
```

```python
x = career_stats['WS']/10
career_stats['Career Value'] = 2/((1/25)+(1/x))
career_stats
```

```python
top3_seasons = stats_df.groupby(['Player', 'Year of Birth']).apply(lambda x: x.sort_values(["WS"], ascending = False)).reset_index(drop=True)
# select top N rows within each continent
top3_seasons = top3_seasons.groupby(['Player', 'Year of Birth']).head(3)
top3_seasons
```

```python
avgtop_3 =  top3_seasons.groupby(['Player', 'Year of Birth'], as_index=False)['WS'].mean().sort_values('WS', ascending=False)
avgtop_3
```

```python
avgtop_3 = avgtop_3.rename(columns={"WS": "Three Best Years"})
avgtop_3
```

```python
career_stats_4of5 = pd.merge(career_stats, avgtop_3, how='outer', on=['Player', 'Year of Birth'])
career_stats_4of5
```

```python
stats_df = stats_df.sort_values(by=['Player', 'key'])
stats_df
```

```python
y = stats_df.groupby(['Player','Year of Birth'])['WS'].rolling(5).sum().reset_index([0,1],drop=True)
y=y.to_frame()
y = y.rename(columns={"WS": "5 year Total WS"})
y
```

```python
stats_df = pd.merge(stats_df, y, left_index=True, right_index=True)
stats_df
```

```python
stats_df = stats_df.drop(columns=['5 year Total'])
stats_df
```

```python
stats_df = stats_df.sort_values(by=['5 year Total WS','Player', 'Year of Birth'], ascending = False)
stats_df
```

```python
Fiveyr_peak =  stats_df.groupby(['Player', 'Year of Birth']).head(1)
Fiveyr_peak
```

```python
Fiveyr_peak = Fiveyr_peak[['Player', 'Year of Birth', '5 year Total WS']]
Fiveyr_peak
```

```python
career_stats_5of5 = pd.merge(career_stats_4of5, Fiveyr_peak, how='outer', on=['Player', 'Year of Birth'])
career_stats_5of5
```

```python
career_stats_5of5['5 year Total WS'] = career_stats_5of5['5 year Total WS']/5
career_stats_5of5
```

```python
career_stats_5of5['Total'] = career_stats_5of5['WS/Season'] + career_stats_5of5['Time line adjusted'] + career_stats_5of5['Career Value'] + career_stats_5of5['Three Best Years'] + career_stats_5of5['5 year Total WS']
career_stats_5of5
```

```python
career_stats_5of5 = career_stats_5of5.sort_values(by=['Total'], ascending = False).reset_index()
career_stats_5of5
```
