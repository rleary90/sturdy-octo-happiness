```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd  
```

```python
ds = pd.read_csv('titanic.csv')
print(ds)
```

```python
%matplotlib inline
Age = ds['Age']
Fare = ds['Fare']
plt.scatter(Age, Fare, edgecolors='r')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Ticket Fare on Titanic By Age')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd  
ds = pd.read_csv('titanic.csv')
legend = ['Fare']
Fare = ds['Fare']
plt.hist([Fare], bins = 10, color=['orange'])
plt.xlabel("Fare")
plt.legend(legend)
plt.title('Titantic Passenger info')
plt.show()
```

```python
legend = ['Age']
Age = ds['Age']
plt.hist([Age], bins = 10, color=['blue'])
plt.xlabel("Age")
plt.legend(legend)
plt.title('Titantic Passenger info')
plt.show()
```

```python
legend = ['Age', 'Fare']
plt.hist(Age, color='blue', bins=np.arange(-10, 40), alpha=.5) 
plt.hist(Fare, color='red', bins=np.arange(-10, 40), alpha=.5)
plt.title('Titantic Passenger Info')
plt.xlabel('Values')
plt.legend(legend)
plt.show()
```

```python
plt.hist(Age, color='blue', density = True, bins=np.arange(-10, 40), alpha=.5) 
plt.hist(Fare, color='red', density = True, bins=np.arange(-10, 40), alpha=.5)
plt.title('Titanic Passenger Info')
plt.xlabel('Values')
plt.legend(legend)

plt.show()
```
