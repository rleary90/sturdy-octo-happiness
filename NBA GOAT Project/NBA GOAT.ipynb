# Create For Loop for all the BAA years, then aba years, then all nba years

```python
yearsBAA = list(range(1947,1950))
yearsABA = list(range(1968,1977))
yearsNBA = list(range(1950,2020))
yearsBAA
```

```python
statsBAA = {}
for yearBAA in yearsBAA:
    # URL page we will scraping (see image above)
    url = "https://www.basketball-reference.com/leagues/BAA_{}_advanced.html".format(yearBAA)
    # this is the HTML from the given URL
    html = urlopen(url)
    soup = BeautifulSoup(html)
    # use findALL() to get the column headers
    soup.findAll('tr', limit=2)
    # use getText()to extract the text we need into a list
    headers = [th.getText() for th in soup.findAll('tr', limit=2)[0].findAll('th')]
    # exclude the first column as we will not need the ranking order from Basketball Reference for the analysis
    headers = headers[1:]
    headers
    # avoid the first header row
    rows = soup.findAll('tr')[1:]
    player_stats = [[td.getText() for td in rows[i].findAll('td')]
            for i in range(len(rows))]
    
    statsBAA[yearBAA] = pd.DataFrame(player_stats, columns = headers)
    statsBAA[yearBAA]['League'] = 'BAA'
```

```python
for k in statsBAA:
    statsBAA[k].insert(0, 'key', [k]*len(statsBAA[k]))

statsBAA_df = pd.concat(statsBAA.values())
display(statsBAA_df)
```

```python
statsABA = {}
for yearABA in yearsABA:    
    # URL page we will scraping (see image above)
    url = "https://www.basketball-reference.com/leagues/ABA_{}_advanced.html".format(yearABA)
    # this is the HTML from the given URL
    html = urlopen(url)
    soup = BeautifulSoup(html)
    # use findALL() to get the column headers
    soup.findAll('tr', limit=2)
    # use getText()to extract the text we need into a list
    headers = [th.getText() for th in soup.findAll('tr', limit=2)[0].findAll('th')]
    # exclude the first column as we will not need the ranking order from Basketball Reference for the analysis
    headers = headers[1:]
    headers
    # avoid the first header row
    rows = soup.findAll('tr')[1:]
    player_stats = [[td.getText() for td in rows[i].findAll('td')]
            for i in range(len(rows))]
    statsABA[yearABA] = pd.DataFrame(player_stats, columns = headers)
    statsABA[yearABA]['League'] = 'ABA'
```

```python
for k in statsABA:
    statsABA[k].insert(0, 'key', [k]*len(statsABA[k]))

statsABA_df = pd.concat(statsABA.values())
display(statsABA_df)
```

```python
statsNBA = {}
for yearNBA in yearsNBA:    
    # URL page we will scraping (see image above)
    url = "https://www.basketball-reference.com/leagues/NBA_{}_advanced.html".format(yearNBA)
    # this is the HTML from the given URL
    html = urlopen(url)
    soup = BeautifulSoup(html)
    # use findALL() to get the column headers
    soup.findAll('tr', limit=2)
    # use getText()to extract the text we need into a list
    headers = [th.getText() for th in soup.findAll('tr', limit=2)[0].findAll('th')]
    # exclude the first column as we will not need the ranking order from Basketball Reference for the analysis
    headers = headers[1:]
    headers
    # avoid the first header row
    rows = soup.findAll('tr')[1:]
    player_stats = [[td.getText() for td in rows[i].findAll('td')]
            for i in range(len(rows))]
    statsNBA[yearNBA] = pd.DataFrame(player_stats, columns = headers)
    statsNBA[yearNBA]['League'] = 'NBA'
```
